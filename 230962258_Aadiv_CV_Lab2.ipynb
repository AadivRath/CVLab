<h3> Example Questions

Log Transformation


```python
img = cv.imread('/home/cvl-5aiml-a2/Downloads/bowzer.jpg',0)
img = img.astype(np.float32)
c = 255/(np.log(1+np.max(img)))
log_transformed = c*np.log(1+img)
log_transformed = np.array(log_transformed, dtype = np.uint8)
cv.imshow('log_transformed.jpg',log_transformed)
cv.waitKey()
cv.destroyAllWindows()
```

Power-Law (Gamma) Transformations

```python
img = cv.imread('/home/cvl-5aiml-a2/Downloads/bowzer.jpg',0)
for gamma in [0.1, 0.5, 1.2, 3.3]:
    gamma_corrected = np.array(255*(img/255)**gamma, dtype = 'uint8')
    cv.imshow('updated image'+str(gamma)+'.jpg', gamma_corrected)
cv.waitKey()
cv.destroyAllWindows()
```

1. Write a program to read an image and perform histogram equalization.


```python
img = cv.imread('/home/cvl-5aiml-a2/Downloads/bowzer.jpg',0)
img.shape 
#cv.imshow('image', img) test to see if it works
#to see the number of channels, notice how there are 3 channels that we can use as a parameter for hist function.

hist = cv.calcHist([img], [0], None, [256],[0,255])
plt.plot(hist)
plt.show()

#For fun, show the equalised histogram
img = cv.equalizeHist(img)
hist = cv.calcHist([img], [0], None, [256],[0,255])
#Hmm interesting, 
'''
the parameters here are image in an array, [0] corresponds to the channel of which you want the histogram calculated,
i.e., we can pass [0], [1] or [2] to calculate histogram of blue, green or red channel respectively. 'None' corresponds
to the cropped area of which we want the histogram (No crops in this case!), [256] stands for the bin counts on the x-axis,
and range represents the range that we want. For most cases, the other parameters will remain the same for us. 
'''

#Plotting the histogram
plt.plot(hist)
plt.show()
```

3. Write a program to Resizing the Image and cropping an image.

```python
image = cv.imread('/home/cvl-5aiml-a2/Downloads/bowzer.jpg')
cv.imshow('Original Image', image)
 
# let's downscale the image using new  width and height
down_width = 300
down_height = 200
down_points = (down_width, down_height)
resized_down = cv.resize(image, down_points, interpolation= cv.INTER_LINEAR)
 
# let's upscale the image using new  width and height
up_width = 600
up_height = 150
up_points = (up_width, up_height)
resized_up = cv.resize(image, up_points, interpolation= cv.INTER_LINEAR)
 
# Display images
cv.imshow('resized down', resized_down)
cv.waitKey()
cv.imshow('donk', resized_up)

cv.waitKey()
cv.destroyAllWindows()
```
